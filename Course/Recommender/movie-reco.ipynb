```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df1 = pd.read_csv('user_ratings.data', sep='\t', names=column_names)
df1.head()
```

```python
movie_titles = pd.read_csv("id_titles")
movie_titles.head()
```

```python
len(df1), len(movie_titles)
```

- merge on item_id using pd.merge() on item_id

```python
df = pd.merge(movie_titles, df1, on='item_id')
df
```

- ratings dataframe by taking a groupby on ratings and mean()

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

- add a column to ratings dataframe of the count of ratings

```python
ratings['count'] = df.groupby('title')['rating'].count()
ratings.head()
ratings.sort_values('count', ascending=False)
```

- create a pivot table df.pivot_table(index='user_id',columns='title',values='rating')

```python
movie_matrix = df.pivot_table(index='user_id',columns='title', values='rating')
movie_matrix.head()
```

- get data from the pivot table for a movie

```python
star_wars = movie_matrix['Star Wars (1977)']
```

```python
star_wars.describe()
```

- use corrwith on the matrix and pass the data of the movie

```python
star_wars_corr = pd.DataFrame(movie_matrix.corrwith(star_wars), columns=['corr'])
# Cold Stat -> Where we cant find corelation between. For Ex: '3 Ninjas: High Noon At Mega Mountain (1998)' below
star_wars_corr.head()
star_wars_corr.sort_values('corr', ascending=False)
```

```python
#star_wars_corr.join(ratings).sort_values('count', ascending=False)
star_wars_corr_count = star_wars_corr.join(ratings)
star_wars_corr_count.head()
```

```python
star_wars_corr_count.sort_values('corr', ascending=False)
```

```python
star_wars_corr_count_min_100 = star_wars_corr_count[star_wars_corr_count['count'] > 100]
star_wars_corr_count_min_100.head()
star_wars_corr_count_min_100.sort_values('corr', ascending=False).head(10)
```
