```python
X = df[['INDUS', 'NOX', 'RM', 'TAX', 'PTRATIO', 'LSTAT']]
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_, lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
print("MAE", mean_absolute_error(y_test, y_pred))
print("MSE", mean_squared_error(y_test, y_pred))
print("RMSE", math.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#Select columns that have correlation between 0.1 and 1
#We should not take the columns having the correlation as 1 because it will result in creating the model which can
#  give expected results based on just this one column and by setting other columns coefficients as 0
corr_choice = np.linspace(0.1, 1, 20)
```

```python
split=[0.4,0.35,0.3,0.25]
```

```python
features = c[(abs(c['target']) > 0.4) & (abs(c['target'] != 1))].index
```

```python
def get_features(each_corr):
    return c[(abs(c['target']) > each_corr) & (abs(c['target'] != 1))].index
```

```python
features = [get_features(f) for f in corr_choice]
```

```python
len(features), type(features), type(features[0]), features[0:20]
```

```python
features = [get_features(f) for f in corr_choice if len(get_features(f)) > 2]
```

```python
len(features), df[features[0]]
```

```python
y = df['target']
def get_mae(f, t):
    X = df[f]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=t, random_state=101)
    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    return mean_squared_error(y_test, y_pred)
```

```python
#get_mae(features[1],split[1])
#type(features), type(split)
results = [ (f , t, get_mae(f,t)) for f in features for t in split]
```
